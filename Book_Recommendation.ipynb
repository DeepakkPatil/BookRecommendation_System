## Popularity Based Recommender System

```python
### we would display top 50 books having ratings>=250 votes
```

```python
ratings.merge(books,on="ISBN")
```

```python
#we only want some specific columns to be the part of our data set
ratings_with_name=ratings.merge(books,on="ISBN")
```

```python
#how many votes are there on each book
ratings_with_name.groupby('Book-Title').count()
```

```python
# we only want ratings so
ratings_with_name.groupby('Book-Title').count()['Book-Rating']
```

```python
# we want this to a new data set
num_rating_df=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
#rename the column Book-Rating to num_ratings
num_rating_df.rename(columns={'Book-Rating':'num_rating'},inplace=True)

# Inplace=True is used to update existing dataset dont create copy
num_rating_df
```

### Average Ratings

```python
avg_rating_df=ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
#rename the column Book-Rating to num_ratings
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)

# Inplace=True is used to update existing dataset dont create copy
avg_rating_df
```

```python
popular_df=num_rating_df.merge(avg_rating_df,on="Book-Title")
popular_df
```

```python
#ONly those books which has num ratings>=250
popular_df=popular_df[popular_df['num_rating']>=250]
```

```python
# sort in descending order to get most rated books
popular_df=popular_df.sort_values('avg_rating',ascending=False)
```

```python
# we need top 50 books
popular_df=popular_df.head(50)
popular_df
```

```python
# now authors name,publisher and image se we merge this
popular_df=popular_df.merge(books,on='Book-Title')
popular_df.shape
```

```python

# But now the book count has incread because same named books 
# are also there multiple times hence remove them
popular_df=popular_df.drop_duplicates('Book-Title')
```

```python
# now check size
popular_df.shape
```

#### Hence only 50 books are there with all details

```python
# now only imp colmns are needed so we use 
popular_df=popular_df[['Book-Title','Book-Author','Image-URL-M','num_rating','avg_rating']]
```

## Collaborative Filtering 

```python
# we need to make table having colmns as users and rows as book 
# each colm would have value of rating of diff book by a user 
ratings_with_name.head(1)
```

```python
# we would only consider those users which have min 200 book ratings
x=ratings_with_name.groupby('User-ID').count()['Book-Rating']>=200
x[x] # boolean indexing 
```

```python
Exp_users=x[x].index
```

```python
filtered_rating=ratings_with_name[ratings_with_name['User-ID'].isin(Exp_users)]
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
y[y] #boolean indexing
```

```python
Famous_books=y[y].index
```

```python
final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(Famous_books)]
```

```python
# Ab hme uss table ke according banana h jisme colm m users
# and rows m book rhe and value book ke aage har user ki di huyi rating rhe
final_ratings.pivot_table(index='Book-Title',columns="User-ID",values="Book-Rating")
```

```python
# lets call this table as pt
pt=final_ratings.pivot_table(index='Book-Title',columns="User-ID",values="Book-Rating")
```

```python
pt.shape
```

```python
# fill NA by 0
pt.fillna(0,inplace=True)
pt
```

```python
pip install sklearn #for installing sklearn
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores=cosine_similarity(pt) #Euclidian distance of each book from the others
```

```python
similarity_scores.shape
```

```python
# we need descending order books in similarity
# but we need index hence first Enumerate kro
```

```python
sorted(list(enumerate(similarity_scores[0])),key=lambda x:x[1],reverse=True) 
#sorting done on the basis of 2nd index
```

```python
# we only need top 5 books excluding itself hence
sorted(list(enumerate(similarity_scores[0])),key=lambda x:x[1],reverse=True)[1:6]
```

```python
def recommend(book_name) :
    #book name se index fetch kro  
    index=np.where(pt.index==book_name)[0][0]
    similar_items= sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    for i in similar_items :
        print(pt.index[i[0]])
        
```

```python
#we have old fucctn which only prints name of book but we need 
#image url and author also so

def recommend_final(book_name) :
    #book name se index fetch kro  
    
    data=[]
    
    index=np.where(pt.index==book_name)[0][0]
    similar_items= sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    for i in similar_items :
        item=[]
        temp=books[books['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```
